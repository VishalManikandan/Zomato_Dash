```python

year = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']


data = {
    'January': [5, 6, 7, 8, 9, 10, 11, 12, 13],
    'February': [3, 4, 5, 2, 6, 7, 8, 9, 10],
    'March': [4, 6, 2, 3, 5, 6, 7, 8, 9],
    'April': [7, 5, 6, 7, 8, 9, 10, 11, 12],
    'May': [8, 9, 10, 11, 12, 11, 10, 9, 8],
    'June': [2, 3, 3, 2, 4, 5, 6, 7, 8],
    'July': [6, 7, 8, 9, 10, 9, 8, 7, 6],
    'August': [5, 4, 6, 7, 8, 9, 10, 11, 12],
    'September': [2, 3, 4, 5, 6, 5, 4, 3, 2],
    'October': [9, 8, 7, 6, 5, 4, 3, 2, 1],
    'November': [5, 6, 7, 8, 9, 10, 11, 12, 13],
    'December': [4, 5, 6, 5, 4, 3, 2, 1, 0],
}

for month in months:
    plt.bar(year, data[month], label=month)


plt.title("Line Plot of Monthly Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Sales")
plt.legend()
plt.xticks(year) 

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
sizes = [5, 6, 7, 8, 9, 10, 11, 12, 13, 10, 9, 8]  # Example sales data
colors = plt.cm.tab10.colors  # Using tab10 color map

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors[:len(sizes)], autopct='%1.1f%%', startangle=140)

plt.axis('equal')  

plt.title("Sales Distribution by Month")

plt.show()
```
